```python
import pandas as pd
from pandas import Series, DataFrame
%pylab inline

from pandas import Series, DataFrame
import pandas as pd

HR_df = pd.read_csv(r'/Users/sanjjj/Desktop/Course/Data Science Programming/Project/data/HRXYZ data.csv')

HR_df
```

```python
# drop rows with mssing values, most rows do not have NA values
HR_df.dropna(inplace=True)

# change attrition to numerical
attrition_mapping = {'Yes': 1, 'No': 0}

# Apply the mapping to the 'Attrition' column
HR_df['Attrition'] = HR_df['Attrition'].map(attrition_mapping)
HR_df

# apply mapping to the 'Gender' column 
gender_mapping = {'Male': 1, 'Female': 0}
HR_df['Gender'] = HR_df['Gender'].map(gender_mapping)

#HR_df

# droppping some columns that we think are not important/not variables of interest
HR_df.drop(['Over18', 'EmployeeCount', 'StandardHours', 'EmployeeID'], axis=1, inplace=True)
HR_df
```

```python
import numpy as np
import scipy.stats as stats
```

```python
#Age
# Convert the DataFrame into a contingency table
contingency_table = pd.crosstab(HR_df['Age'], HR_df['Attrition'])

# Perform the Chi-square test
chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

print(f"Chi-square test statistic: {chi2_stat}")
print(f"P-value: {p_val}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
```

```python
# If the p-value is less than the significance level (e.g., 0.05), you can reject the null hypothesis (H0) and conclude that there is a statistically significant association between the categorical variables. In other words, the observed differences in the frequencies are not likely to have occurred by chance alone, and there is evidence of a relationship between the variables.

# If the p-value is greater than or equal to the significance level, you fail to reject the null hypothesis. This suggests that there is no statistically significant association between the categorical variables, and any observed differences in the frequencies could be due to chance.

# In summary, a small p-value indicates a significant association between the categorical variables, while a large p-value suggests no significant association. The Chi-square test does not provide information on the strength or direction of the association; it only determines if an association exists or not
```
